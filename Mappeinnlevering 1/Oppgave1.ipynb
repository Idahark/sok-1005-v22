```r
library(tidyverse)
library(readr)
library(ggplot2)
library(data.table)
library(zoo)

lower <- fread("https://www.nsstc.uah.edu/data/msu/v6.0/tls/uahncdc_ls_6.0.txt")

lower2 = rollmean(lower, 13)

df_lower2 <- as.data.frame(lower2)

df_lower2_1 <-
  df_lower2 %>%
  select(Year, Globe)

df_lower2_1 %>%
  ggplot(aes(x=Year, y=Globe))+
  geom_line()+
  labs(title = "Globale temperaturer, nedre troposfære",
       x = "År",
       y = "Gjennomsnittstemperatur")+
  theme_bw()
```
