```r
## Oppgave 2 Mappe 1

library(tidyverse)
library(readr)
library(ggplot2)
library(data.table)
library(zoo)
library(dplyr)
```

```r
low_trop <- fread("https://www.nsstc.uah.edu/data/msu/v6.0/tls/uahncdc_ls_6.0.txt")

mid_trop <- fread("https://www.nsstc.uah.edu/data/msu/v6.0/tmt/uahncdc_mt_6.0.txt")

mid_trop_ren <- mid_trop[-c(518),]

trop <- fread("https://www.nsstc.uah.edu/data/msu/v6.0/ttp/uahncdc_tp_6.0.txt")

trop_ren <- trop[-c(518),]

low_strop <- fread("https://www.nsstc.uah.edu/data/msu/v6.0/tls/uahncdc_ls_6.0.txt")
```

```r
##Gjøre dataen leselig

low_trop$year <- as.numeric(low_trop$Year)
low_trop$roll <- as.numeric(low_trop$NoPol)

kanskje3 <-
  low_trop %>%
  select(year, roll)

low_trop_num <- rollmean(kanskje3, 13)

##2

trop_ren$roll <- as.numeric(trop_ren$NoPol)
trop_ren$year <- as.numeric(trop_ren$Year)

kanskje <-
  trop_ren %>%
  select(year, roll)

trop_ren_num <- rollmean(kanskje, 13)

##3

mid_trop_ren$year <- as.numeric(mid_trop_ren$Year)
mid_trop_ren$roll <- as.numeric(mid_trop_ren$NoPol)

kanskje1 <-
  mid_trop_ren %>%
  select(year, roll)

mid_trop_ren_num <- rollmean(kanskje1, 13)

##4

low_strop$year <- as.numeric(low_strop$Year)
low_strop$roll <- as.numeric(low_strop$NoPol)

kanskje2 <-
  low_strop %>%
  select(year, roll)

low_strop_num <- rollmean(kanskje2, 13)
```

```r
## bearbeidet dataframe

df_low_trop_roll <- as.data.frame(low_trop_num)

df_mid_trop_roll <- as.data.frame(mid_trop_ren_num)

df_trop_roll <- as.data.frame(trop_ren_num)

df_low_strop <- as.data.frame(low_strop_num)

##sammenslåing av data

total1 <- rbind(df_low_trop_roll, df_mid_trop_roll)

total2 <- rbind(df_trop_roll, df_low_strop)

total3 <- rbind(total1, total2)

gjnst <-
  total3 %>%
  summarise(mean(roll))
```

```r
##Plot

sammenlign <- ggplot(NULL, aes(x=year, y=roll)) +
  geom_line(data=df_low_trop_roll, col="red")+
  geom_line(data=df_mid_trop_roll, col="purple")+
  geom_line(data=df_trop_roll, col="blue")+
  geom_line(data=df_low_strop, col="pink")+
  geom_hline(yintercept = 0.0654491, col = "black")+
  labs(title = "Gjennomsnittstemperatur  i de forskjellige spherene, 60-90 grader nord", 
       caption = "Rød=Nedre Troposfære, Lila=midt troposfære, Blå=Tropopause, Rosa=Nedre tropopause, Svart=Gjennomsnitt",
       x = "År",
       y = "Temperatur")
sammenlign
```
